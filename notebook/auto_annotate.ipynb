```python
import re

stopwords_vi = set([
    "và", "của", "các", "là", "cho", "trong", "với", "cũng",
])

# preprocessing
def preprocess_text_vi(text, stopwords = stopwords_vi):
    text = re.sub(r'[^\w\s]', '', text)
    text = text.lower()
    tokens = text.split()
    tokens = [word for word in tokens if word not in stopwords]
    preprocessed_text = ' '.join(tokens)
    return preprocessed_text
```

```python
from sklearn.feature_extraction.text import CountVectorizer

def top_50_most_common_words(df):
    # Lấy danh sách các câu từ cột 'title'
    titles = df.tolist()
    
    # Khởi tạo và fit CountVectorizer
    vectorizer = CountVectorizer()
    X = vectorizer.fit_transform(titles)
    
    # Tính toán số lần xuất hiện của mỗi từ
    word_frequencies = X.sum(axis=0)
    word_frequencies = [(word, word_frequencies[0, idx]) for word, idx in vectorizer.vocabulary_.items()]
    
    # Sắp xếp theo số lần xuất hiện giảm dần
    sorted_word_frequencies = sorted(word_frequencies, key=lambda x: x[1], reverse=True)
    
    # Lấy top 30 từ xuất hiện nhiều nhất
    top_50_words = sorted_word_frequencies[:50]
    
    return top_50_words
```

```python
# import re
def word_count(df, n):
    word_count = {}
    for row in df.values:
        # text = re.sub(r'[^\w\s]', '', row)  
        # text = text.lower()
        # tokens = text.split()
        tokens = row.split()
        for i in range(len(tokens) - n + 1):
            ngram = " ".join(tokens[i:i+n])
            if ngram not in word_count:
                word_count[ngram] = 1
            else:
                word_count[ngram] += 1
    return word_count
```

```python
gd_acceptable_list = ['học sinh', 'tuyển sinh', 'tốt nghiệp', 'thi tốt', 'mùa thi', 'trường thpt', 'thpt chuyên', 'trường đh', 'giáo viên', 'lớp 6', 'bộ gdđt', 'trường quốc tế']
kt_acceptable_list = ['thuế', 'ngân hàng', 'giá vàng', 'giá usd', 'lãi suất']
cng_acceptable_list = []
sk_acceptable_list = ['sức khỏe', 'ung thư', 'điều trị', 'thực phẩm']
xe_acceptable_list = ['ô tô']
```

```python
def check_title(text):
    ans = []
    for word in gd_acceptable_list:
        if word in text:
            ans.append('giao-duc')
            break
    for word in kt_acceptable_list:
        if word in text:
            ans.append('kinh-te')
            break
    for word in cng_acceptable_list:
        if word in text:
            ans.append('cong-nghe-game')
            break
    for word in sk_acceptable_list:
        if word in text:
            ans.append('suc-khoe')
            break
    for word in xe_acceptable_list:
        if word in text:
            ans.append('xe')
            break
    if (len(ans) == 1):
        return ans[0]
    else:
        return None
    
df['annotated_genre'] = df['title'].apply(check_title)
df[df['annotated_genre'] == None]
```
